---

### **Introducción**


El presente informe tiene como propósito apoyar al señor Juan en la toma de decisión estratégica sobre cual de las cuatro tiendas (tienda, tienda2, tienda3 o tienda4) debería considerar vender para iniciar un nuevo proyecto. La decisión se basa en un análisis exhaustivo de datos de ventas, desempeño logístico y satisfacción del cliente. Se han evaluado cinco dimensiones clave: ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio, productos más y menos vendidos, y el coste de envío promedio.

---

### **Desarrollo del Análisis**

1.   **Ingresos Totales por Tienda**

Los ingresos generados por cada tienda muestran una clara diferencia en el rendimiento financiero:

* **Tienda**: $1.150.880.400

* **Tienda2**: $1.116.343.500

* **Tienda3**: $1.098.019.600

* **Tienda4**: $1.038.375.700

**Conclusión: Tienda4** es la que menos ingresos genera.
    

2.   **Categorías de Productos Más y Menos Vendidos**

Todas las tiendas tienen una distribución de ventas bastante similar. Las categorías más vendidas en general son:

* **Muebles** (Mayor volumen en todas)
* **Electrónicos**
* **Juguetes**

Las menos vendidas suelen ser:

* **Instrumentos musicales**
* **Libros**
* **Articulos para el hogar**

**Conclusión:** No hay diferencias relevantes en este aspecto entre las tiendas; todas venden productos similares en proporciones parecidas.


3. **Calificaciones Promedio de Clientes**

* **Tienda**: 3.98
* **Tienda2**: 4.04
* **Tienda3**: 4.05
* **Tienda4**: 3.99

**Conclusión: Tienda3** obtiene la mejor valoración de los clientes, lo cual es un indicio de una experiencia de compra satisfactoria. Mientras que la **Tienda**es la peor valorada. Aun así, la diferencia entre todas es pequeña.

4. **Productos Más y Menos Vendidos**

Las tiendas comparten una gran parte del portafolio, pero **Tienda4** presenta un menor volumen en ventas de varios productos clave. Ejemplos:

* El producto más vendido en **Tienda** (Microondas) registró 60 unidades.

* En **Tienda4**, los productos más vendidos (Cama box, Cubertería) no superan las 62 unidades, pero muchos productos muestran ventas estancadas o menores que en otras tiendas.

**Conclusión: Tienda4** tiene menor rotación de productos, lo que puede sugerir baja demanda o una ubicación menos favorable.


5. **Coste de Envio Promedio**

* **Tienda**: $26.019

* **Tienda2**: $2.216

* **Tienda3**: $24.805

* **Tienda4**: $23.459 (más bajo)

**Tienda4** ofrece el coste de envio más bajo lo que puede ser atractivo para consumidores sensibles al precio final, aunque no compensa sus debilidades en otros indicadores.

---

### **Conclusión y Recomendación**

Tras analizar todos los factores, se concluye que la tienda que el Sr. Juan debería vender es la **Tienda4**. Esta recomendación se fundamenta en:

* Menores ingresos totales.

* Peor desempeño en ventas de productos clave.

* Ligera desventaja en la calificación promedio frente a otras tiendas.

* Poca diferencia significativa en las categorías vendidas.

Aunque su costo de envío es el más bajo, esta fortaleza no es suficiente para contrarrestar sus debilidades. Vender esta tienda representa una menor pérdida de valor estratégico para Alura Store y permitirá al Sr. Juan capitalizar un activo con bajo rendimiento para invertir en un nuevo emprendimiento.
